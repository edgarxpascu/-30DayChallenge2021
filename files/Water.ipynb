```python
%matplotlib inline

import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from cartopy import crs as ccrs
```

```python
rh = gpd.read_file("red_hidrografica_250k/red_hidrografica_250k.shp")
```

Selecci√≥n por orden en la red

```python
O7 = rh.query("ORDEN== 7 ")
O6 = rh.query("ORDEN== 6 ")
O5 = rh.query("ORDEN== 5 ")
O4 = rh.query("ORDEN== 4 ")
O3 = rh.query("ORDEN== 3 ")
O2 = rh.query("ORDEN== 2 ")
O1 = rh.query("ORDEN== 1 ")
O0 = rh.query("ORDEN== 0 ")
```

```python
f, ax = plt.subplots(1,figsize=(20, 16))

# Removes ticks and labels for lat/lon
ax.tick_params(
    axis='both', bottom=False, left=False,         
    labelbottom=False, labelleft=False) 

O7.plot(ax = ax, linewidth = 1.4, color = "#CAF0F8", zorder = 9)
O6.plot(ax = ax, linewidth = 1.2, color = "#ADE8F4", zorder = 8)
O5.plot(ax = ax, linewidth = 1.0, color = "#90E0EF", zorder = 7)
O4.plot(ax = ax, linewidth = 0.8, color = "#48CAE4", zorder = 6)
O3.plot(ax = ax, linewidth = 0.6, color = "#00B4D8", zorder = 5)
O2.plot(ax = ax, linewidth = 0.4, color = "#0096C7", zorder = 4)
O1.plot(ax = ax, linewidth = 0.2, color = "#0077B6", zorder = 3)
O0.plot(ax = ax, linewidth = 0.2, color = "#023E8A", zorder = 2)

ax.set_title('Red Hidrogr√°fica de M√©xico', fontsize=25, pad=10)
ax.set_facecolor('black')

plt.show()
```
